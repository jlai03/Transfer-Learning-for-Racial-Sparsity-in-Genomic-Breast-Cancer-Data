```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Open Processed Clinical Data
datapath = "../data/processed_labels/clinical.tsv"
df = pd.read_csv(datapath, sep='\t')
df
```

```python
# Count unique values in the 'Primary Diagnosis' column
value_counts = df['primary_diagnosis'].value_counts()
value_counts
```

```python
# Plotting the bar graph
plt.figure(figsize=(8, 6))
value_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Prognoses')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
# Count unique values in the 'Race' column
race_counts = df['race'].value_counts()
race_counts
```

```python
# Plotting the bar graph
plt.figure(figsize=(4, 5))
race_counts.plot(kind='bar', color='pink')
plt.title('Distribution of Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
```
