```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Open Processed Clinical Data
datapath = "../data/processed_labels/clinical.tsv"
df = pd.read_csv(datapath, sep='\t')
df
```

```python
# Count unique values in the 'Category' column
value_counts = df['primary_diagnosis'].value_counts()
value_counts
```

```python
# Plotting the bar graph
plt.figure(figsize=(8, 6))
value_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Unique Values in Category Column')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
